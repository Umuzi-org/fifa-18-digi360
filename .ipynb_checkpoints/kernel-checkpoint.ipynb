**This dataset contains information about Players from the EA Sports game, FIFA 18. We will take this data, clean it, analyse it, and gain insights from it. First let's load the data.**

```python
odf = pd.read_csv("CompleteDataset.csv", low_memory=False) # odf = original dataframe with complete dataset
odf.head()
```

**Now, let's inspect it...**

```python
def showDetails(df):
    print("-------------------------------------------------------------------------------------------------------------------")
    print('{:>35}'.format("Shape of dataframe:") + '{:>12}'.format(str(df.shape)))
    containsNulls = "Yes" if df.isnull().any().any() else "No"
    print("Does dataframe contain null values: " + containsNulls)
    null_columns = df.columns[df.isnull().any()]
    print("Number of columns with null values: " + str(df[null_columns].isnull().any().sum()))
    null_rows = df[df.isnull().any(axis=1)][null_columns]
    print("Number of records with null values: " + str(len(null_rows)))
    print('{:>35}'.format("Percentage of null records:") + '{:>6.2f}'.format(len(null_rows) / len(df) * 100) + "%")
    print("-------------------------------------------------------------------------------------------------------------------")

showDetails(odf)
```

**Our dataframe contains 17 981 records, with 75 columns. There are 27 columns and 2 235 records with null values, and these null records account for 12.43% of the total number of records .**

**Let's inspect these null records to see if we can find the source of the missing data...**

```python
nv_df = odf[odf.isnull().any(axis=1)] # nv_df ==> null value dataframe
showDetails(nv_df)
nv_df.head()
```

**A quick inspection of the null value records in our dataframe show that those null value records are for goalkeepers, mostly. This is probably because goalkeepers don't have certain characteristics that are only relevent to other positions recorded.**

**We are now going to create a new working dataframe with columns that are relevent to our purposes of applying simple regression on the data.**

```python
wdf = odf[['Name', 'Age', 'Overall', 'Potential', 'Value', 'Wage', 'Preferred Positions']]
showDetails(wdf)
```

**It turns out by choosing these columns, we dropped all the null-values from the dataframe. So now we still have 17 981 records, just with 7 columns**

```python
wdf.head()
```

**In order to apply regression with the columns "Value" and "Wage", we have to convert their string values to numeric types. Here's a function to do that...**

```python
def toFloat(string):
    """Function to convert Wage and Value strings to floats"""
    string = string.strip(" ")
    if string[-1] == 'M':
        return float(string[1:-1]) * 1000000
    elif string[-1] == 'K':
        return float(string[1:-1]) * 1000
    else:
        return float(string[1:])
```

```python
wdf = odf[['Name', 'Age', 'Overall', 'Potential', 'Value', 'Wage', 'Preferred Positions']]
wdf['Value'] = [toFloat(value) for value in wdf['Value']]
wdf['Wage'] = [toFloat(wage) for wage in wdf['Wage']]
wdf.head()
```

**Now the "Value" and "Wage' columns are floats, so we can run a regression on them. Let's first define a function for plotting two columns and optionaly plotting a trend line...**

```python
def myplot(x, y, labelx, labely):
    ax = plt.scatter(x, y)
    z = np.polyfit(x, y, 1)
    p = np.poly1d(z)
    plt.set_xlabel(labelx)
    plt.set_ylabel(labely)
    plt.title('The Title')
    plt.plot(x, p(x), "red")
    plt.show()
```

**A good insight to gather would be the relationship between a player's overall rating and their value.**

```python
myplot(x=wdf['Overall'], y=wdf['Value'])
```

**Our data seems to be producing a skewed graph, and the trend-line doesn't fit quite right. Let's apply a log-transform on the value column to make it less skewed.**

```python
wdf['LoggedValue'] = np.log(wdf['Value'])
wdf['LoggedValue'].tail()
```

```python
myplot(x=wdf['Overall'], y=wdf['LoggedValue'])
```

**Our graph seems more linear now that we've applied a log to the 'Value' column. Let's now apply regression stats on it .**

```python
from matplotlib.pyplot import subplots, show
a = [1, 2, 3, 4, 5]
b = [2, 3, 2, 3, 2]
fig, ax = subplots()
ax.plot(a, b)
ax.set_xlabel("Time")
ax.set_ylabel("Speed")
show()
```
